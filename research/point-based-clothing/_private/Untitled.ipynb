```python
from tqdm import tqdm
import imageio

gif = imageio.get_reader('assets/images/vton1.gif')
gif1 = imageio.get_reader('assets/images/vton2.gif')
gif2 = imageio.get_reader('assets/images/vton3.gif')

print(len(gif), len(gif1), len(gif2))
length = min(len(gif), min(len(gif1), len(gif2)))

images = []
for i in tqdm(range(length)):
    f, f1, f2 = gif.get_data(i), gif1.get_data(i), gif2.get_data(i)
    
    f = cv2.resize(f, (799, 1103))
    ff = np.hstack((f, f1, f2))
    
    images.append(ff)

imageio.mimsave('./vton_all.gif', images)

# import matplotlib.pyplot as plt
# plt.figure(figsize=(10,5))
# plt.imshow(ff)
# plt.show()
```
